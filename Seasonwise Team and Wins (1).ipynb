```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

#import seaborn as sns

%matplotlib inline  
```

We are looking at how each team has performed every season they have participated in.

```python
data = pd.read_csv("matches.csv")
data.head()
```

```python
seasonwise_win = pd.DataFrame(data.groupby(['season','winner']).size())
seasonwise_win.columns = ['Count of Matches Won']
seasonwise_win.reset_index(inplace = True)
seasonwise_win = seasonwise_win.sort_values(by = ['season','Count of Matches Won'],ascending = False)
seasonwise_win.head()
```

We now have a DataFrame who entry are the season, the team and the number of matches won by them in a particular year. This was then sorted in descending order of year and count. Now we will plot the graph of each season performance by each team and see how each team has performed every year.

```python

colors = cm.rainbow(np.linspace(0, 1, len(seasonwise_win['winner'].unique())))
for title, group in seasonwise_win.groupby('season'):
    group.plot(x='winner', y='Count of Matches Won', title=title,kind = 'bar',color = colors)
    plt.legend(loc=1, fontsize = 'small')
```

So above are the graphs showing which team has won how many games each season.Using this we can compare how each of the teams performed in the same season as compared to other season. Now we will see how each team has performed in all the seasons

```python
colors = cm.rainbow(np.linspace(0, 1, len(seasonwise_win['season'].unique())))
for title, group in seasonwise_win.groupby('winner'):
    group.plot(x='season', y='Count of Matches Won', title=title,kind = 'bar',color = colors)
    plt.legend(loc=1, fontsize = 'small')
```

This above figures now shows how each of the team has performed in all the seasons of IPL they have participated in.
