```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('matches.csv')
data.head()
```

We will look at how the teams have worn by runs and wickets

# Win Margin by Runs

```python
runs = data[['season','win_by_runs']]
runs.reset_index(inplace = True)
runs = runs[runs['win_by_runs']!=0]
runs.head()
```

```python
sns.set(font_scale=2)
g = sns.lmplot(x = 'index',y = 'win_by_runs', data = runs,fit_reg = False,hue = 'season')
g.fig.set_size_inches(15,10)
plt.xlabel('Match Number',fontsize = 20)
plt.ylabel('Win Margin(Runs)',fontsize = 20)
plt.title('Win Margin by runs of each Match',fontsize = 20)
plt.show()
```

This graph tells us what the general win margin by runs is for every match till now.

# Win Margin by Runs Across Season

```python
runs_season = data[['season','win_by_runs','winner']]
runs_season = runs_season[runs_season['win_by_runs']!=0]
runs_season.head()
```

```python
sns.set(font_scale=2)
plt.rcParams['figure.figsize'] = 15,10
g = sns.swarmplot(x ='season', y = 'win_by_runs',hue = 'winner', data = runs_season,size = 10)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.xlabel('Season',fontsize = 20)
plt.ylabel('Win Margin(Runs)',fontsize = 20)
plt.title('Win Margin Across all Seasons')
plt.show()
```

This shows a swarmplot of the win margin every season across all teams

Now we will look at the same data but now we will divide it by each team

# Win Margin by Wickets

```python
wicket = data[['winner','win_by_wickets']]
wicket.reset_index(inplace = True)
wicket = wicket[wicket['win_by_wickets']!=0]
wicket.head()
```

```python
sns.set(font_scale=1.5)
g = sns.lmplot(x = 'index',y = 'win_by_wickets', data = wicket,fit_reg = False,hue = 'winner',size = 12)
g.fig.set_size_inches(15,10)
plt.xlabel('Match Number',fontsize = 20)
plt.ylabel('Win Margin(Wicket)',fontsize = 20)
plt.title('Win Margin by wickets of each Match',fontsize = 20)
plt.show()
```

This graph tells us what the general win margin by runs is for every match till now.

# Win Margin by Wickets Across Season

```python
wickets_season = data[['season','win_by_wickets','winner']]
wickets_season = wickets_season[wickets_season['win_by_wickets']!=0]
wickets_season.head()
```

```python
sns.set(font_scale=1.5)
plt.rcParams['figure.figsize'] = 20,15
g = sns.swarmplot(x ='season', y = 'win_by_wickets',hue = 'winner', data = wickets_season,size = 10)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.xlabel('Season',fontsize = 20)
plt.ylabel('Win Margin(Wickets)',fontsize = 20)
plt.title('Win Margin Across all Seasons')
plt.show()
```
