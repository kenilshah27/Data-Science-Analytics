```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('matches.csv')
data.head()
```

Here we will be looking at the stadiums that were used most number of times and the cities which hosted the most number of matches.

```python
city_count = data.groupby(['city']).size()
city_count = city_count.sort_values(ascending = False)
city_count.head()
```

```python
print('Total cities played at:',data['city'].nunique())
```

This shows top 5 cities that has hosted the maximum number of matches in IPL from 2008-2018. This includes all the finals and playsoff hosted by the cities as well apart from the league matches.

```python
plt.figure(figsize = (20,20))
city_count.plot(kind = 'bar')
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.title('Number of Matches at Each City in IPL')
plt.show()
```

Here we can see that the cities which are the home for the franchises playing in the IPL has hosted the most number of matches. With the other cities getting few of the matches. Now we will be plotting only the top 10 cities in the list.

```python
city_count[0:9].plot(kind = 'barh') # we start from 0 as index in python starts from 0
plt.xlabel('Top 10 City')
plt.ylabel('Number of Matches')
plt.title('Number of Matches at Top 10 Cities in IPL')
plt.show()
```

Now we will look at what are the venues that were used the most number of times in all the seasons

```python
venue_count = data.groupby(['venue']).size()
venue_count = venue_count.sort_values(ascending = False)
venue_count.head()
```

```python
print('Total venues played at:',data['venue'].nunique())
```

```python
plt.figure(figsize = (20,20))
venue_count.plot(kind = 'bar')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.title('Number of Matches at Each Venue in IPL')
plt.show()
```

Now we will plot the top 10 venues that has hosted the most number of matches

```python
venue_count[0:9].plot(kind = 'barh') # we start from 0 as index in python starts from 0
plt.xlabel('Top 10 Venue')
plt.ylabel('Number of Matches')
plt.title('Number of Matches at Top 10 Venues in IPL')
plt.show()
```

So this file was about the cities and the venues that has hosted the most number of matches in IPL till 2018.
